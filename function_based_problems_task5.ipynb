# Problem 1: Social Media – Post Engagement Analyzer
## Problem Statement
* Create a function to analyze engagement on a social media post.

## Rules
1. Calculate total likes
2. If Total Likes ≥ 1000 → Viral Post
3. Otherwise → Normal Engagement
## Real-Time Use
* Social media analytics platforms
* Marketing dashboards
## Hint
Use function
Use loop for calculations
## Expected Output Format
Total Likes: 1050
Post Status: Viral Post


```python
def analyze_post_engagement(likes_list):
    """
    Calculates total likes and determines post status.
    """
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example Test
likes = [200, 300, 250, 300]
total, status = analyze_post_engagement(likes)

print("Total Likes:", total)
print("Post Status:", status)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.
## Rules
1. Calculate average rainfall
2. Average ≥ required level → Adequate Rainfall
3. Otherwise → Inadequate Rainfall
## Real-Time Use
* Climate monitoring systems
* Irrigation planning tools
## Hint
Use function
Use loop and math operations
## Expected Output Format
Average Rainfall: 72
Rainfall Status: Adequate Rainfall


```python
def rainfall_checker(rainfall_list, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Example Test
avg, status = rainfall_checker([60, 80, 70, 78], 70)

print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

## Problem 4: Social Media – Duplicate Account Detection
# Problem Statement
Create a function to detect duplicate usernames in a system.
## Rules
1. If duplicates exist → Flag accounts
2. Else → Accounts are unique
## Real-Time Use
Platform moderation
Fraud detection systems
## Hint
Use function
Use set data structure
## Expected Output Format
Duplicate Accounts Found: Yes


```python
def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames using set.
    """
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        return "Yes"
    else:
        return "No"


# Example Test
result = detect_duplicate_accounts(["ashu", "rahul", "ashu"])
print("Duplicate Accounts Found:", result)
```

# Problem 6: Agriculture – Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.
## Rules
1. Price > ₹2000 → Premium Crop
2. Else → Regular Crop
## Real-Time Use
Market analytics
Export pricing systems
## Hint
Use function
Use list filtering logic
## Expected Output Format
Premium Crops: [2500, 3200]


```python
def filter_premium_crops(prices):
    """
    Filters crops above ₹2000.
    """
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


# Example Test
premium = filter_premium_crops([1500, 2500, 3200, 1800])
print("Premium Crops:", premium)
```

# Problem 9: E-Learning – Student Attendance Eligibility System
## Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
## Rules
1. Attendance ≥ 75% → Eligible
2. Attendance < 75% → Not Eligible
# Real-Time Use
Learning management systems
Internship tracking platforms
## Hint
Use function
Use list and loop
## Expected Output Format
Attendance Percentage: 80.0
Exam Eligibility: Eligible



```python
def attendance_checker(attendance_list):
    """
    Calculates attendance percentage and checks eligibility.
    """
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Example Test
percentage, eligibility = attendance_checker([1,1,0,1,1])
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

# Problem 10: Smart Electricity Bill Analyzer
## Problem Statement
* You are building a smart electricity billing system for residential users.
* Create a function that calculates the electricity bill and classifies usage level based on units consumed.
## Billing Rules
1. First 100 units → ₹3 per unit
2. Next 100 units (101–200) → ₹5 per unit
3. Above 200 units → ₹7 per unit
4. Usage Classification
5. Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
## Real-Time Use
Smart meter systems
Electricity board billing software
Energy consumption analytics
## Hint
Use function to calculate bill
Use conditional logic for slabs
Use return values (bill and usage status)


```python
def electricity_bill(units):
    """
    Calculates electricity bill and usage classification.
    """
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example Test
bill, usage = electricity_bill(250)

print("Total Bill:", bill)
print("Usage Classification:", usage)
```
