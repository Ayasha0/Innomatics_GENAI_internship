#Problem 1: Mobile Recharge Validation System
## Problem Statement
Create a function to validate a mobile recharge request.
## Rules:
1. Recharge amount must be ≥ ₹50
2. Valid plans: 199, 299, 399, 599
3. User should be allowed to retry if invalid
## Real-Time Use
* Telecom recharge apps
* Payment gateways
## Hint
1. Use function for validation
2. Use while loop for retry

```python
# Function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount >= 50 and amount in valid_plans:
        return True
    else:
        return False


# Retry until valid
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid Recharge! Try Again")
```

# Problem 2: Inventory Reorder Alert System
## Problem Statement
Given a dictionary of products and stock quantities:
1. Print Reorder Alert if stock < 15
2. Print Stock OK otherwise
## Real-Time Use
* Warehouse systems
* Retail management
##Hint
1. Use function
2. Use for loop on dictionary

```python
# Function to check stock
def check_inventory(stock):

    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Product stock
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

# Problem 3: Student Result Processing System
## Problem Statement
Create a function that:
* Accepts a list of student marks
* Calculates average marks
* Prints Pass if average ≥ 50 else Fail
## Real-Time Use
* Online exam systems
* Evaluation portals
## Hint
1. Use loop to calculate total
2. Return result from function


```python
# Function to process result
def check_result(marks):

    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Student marks
marks = [55, 60, 45, 70, 50]

result = check_result(marks)

print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option
## Problem Statement
Create a cab fare calculator:
1. Base fare = ₹50
2. ₹12 per km
3. Peak hour → 25% extra
4. Allow retry using while loop
## Real-Time Use
* Ride-hailing apps
* Transport systems
## Hint
1. Function for fare calculation
2. While loop for retry


```python
# Function to calculate fare
def calculate_fare(km, peak):

    base_fare = 50
    per_km = 12

    fare = base_fare + (km * per_km)

    if peak == "yes":
        fare = fare * 1.25

    return fare


# Retry system
while True:

    km = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak)

    print("Total Fare: ₹", round(total_fare, 2))

    again = input("Calculate again? (yes/no): ").lower()

    if again != "yes":
        break
```

# Problem 5: Employee Attendance Eligibility Checker
## Problem Statement
Create a function that:
1. Accepts attendance list ("P", "A")
2. Calculates attendance percentage
3. Returns Eligible if ≥ 75% else Not Eligible
## Real-Time Use
* HR payroll systems
* Internship tracking
## Hint
* Use loop to count present days
* Use return statement


```python
# Function to check attendance
def check_attendance(attendance):

    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Attendance data
attendance = ["P", "P", "A", "P", "P", "A", "P"]

status = check_attendance(attendance)

print("Status:", status)
```

# Problem 6: Password Strength Checker
## Problem Statement
Create a function to check password strength.
## Rules:
1. Minimum length = 8
2. Must contain at least one digit
3. Must contain at least one special character (@#$)
## Real-Time Use
* Login systems
* Security validation
## Hint
1. Use loop over string
2. Use conditions inside function


```python
# Function to check password strength
def check_password(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check each character
    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_chars:
            has_special = True

    # Final validation
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Input password
pwd = input("Enter password: ")

result = check_password(pwd)

print(result)
```
