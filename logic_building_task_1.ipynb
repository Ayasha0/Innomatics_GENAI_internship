```python
# 1. User Login Check
print("1: User Login Check")

username = "admin"
password = "1234"

input_username = input("Enter username: ")
input_password = input("Enter password: ")

if input_username == username and input_password == password:
    print("Login Successful\n")
else:
    print("Invalid Credentials\n")



# 2. Pass / Fail Analyzer
print("2: Pass / Fail Analyzer")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)



# 3. Simple Data Cleaner
print("3: Simple Data Cleaner")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names List:", cleaned_names)



# 4. Message Length Analyzer
print("4: Message Length Analyzer")

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Long Message (Flagged)")
    else:
        print("Status: Normal Message")



# 5. Error Message Detector
print("5: Error Message Detector")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR Messages:", error_count)


# End of Program
print("task 1 completed successfully.")
```
